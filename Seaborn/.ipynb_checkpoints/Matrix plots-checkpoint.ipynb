```python
import seaborn as sbn
```

```python
tips = sbn.load_dataset('tips')
```

```python
flights = sbn.load_dataset('flights')
```

```python
tips.he
```

```python
flights.head()
```

```python
# создаем таблицу в виде матрицы чтобы работать с матричными функциями
tips_corr = tips.corr()
```

```python
tips_corr
```

```python
sbn.heatmap(tips_corr, annot=True)
```

```python
flight_pivot = flights.pivot_table(index='month', columns='year', values='passengers')
```

```python
flight_pivot
```

```python
sbn.heatmap(flight_pivot, cmap='coolwarm', linecolor='white', linewidths=3)
# cmap - меняет цветовую палитру
```

```python
sbn.heatmap(flight_pivot)
```

```python
sbn.clustermap(flight_pivot, standard_scale=1)
# standard_scale меняет значения шкалы
```
