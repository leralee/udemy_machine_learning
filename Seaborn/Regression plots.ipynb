```python
import seaborn as sbn
```

```python
tips = sbn.load_dataset('tips')
```

```python
tips.head()
```

```python
sbn.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['>', 'v'], scatter_kws={'s':100})
```

```python
sbn.lmplot(x='total_bill', y='tip', data=tips, col='day', hue='sex', aspect=0.3, height=7)
```
