```python
import seaborn as sbn
```

```python
tips = sbn.load_dataset('tips')
```

```python
tips.head()
```

```python
sbn.distplot(tips['total_bill'], kde=False)
```

```python
sbn.barplot(x='sex', y='total_bill', data=tips)
# среднее значение счетов по категорийным значениям
# не работает если передать два аргумента категорийных столбцов
```

```python
import numpy as np
sbn.barplot(x='sex', y='total_bill', data=tips, estimator=np.std)
# estimator показывает стандартное отклонение по категориям
# не работает если передать два аргумента категорийных столбцов
```

```python
sbn.countplot(x='sex', data=tips)
```

```python
sbn.boxplot(x='day', y='total_bill', data=tips)
# не работает если передать два аргумента категорийных столбцов
# можно передедать два численных, но график некрасивый
```

```python
sbn.boxplot(x='day', y='total_bill', data=tips, hue='smoker')
# hue определяет дополнительный параметр столбца и показывает больше информации о данных
# не работает если передать два аргумента категорийных столбцов
# можно передедать два численных, но график некрасивый
```

```python
sbn.violinplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
# показывает всю информацию о точках
```

```python
sbn.stripplot(x='day', y='total_bill', data=tips, hue='sex', dodge=True)
```

```python
sbn.swarmplot(x='day', y='total_bill', data=tips)
# комбинирование stripplot a violinplot
```

```python
# можно комбинировать
sbn.violinplot(x='day', y='total_bill', data=tips)
sbn.swarmplot(x='day', y='total_bill', data=tips, color='white')
```

```python
sbn.catplot(x='day', y='total_bill', data=tips, kind='bar')
# универсальная функция, отображает график в зависимости от параметра kind
```
