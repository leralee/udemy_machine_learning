```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
dots = sns.load_dataset('dots')
```

```python
dots.head()
```

```python
type(dots)
```

```python
sns.distplot(dots['firing_rate'], kde=False, bins=30)
# kde - линия (kernel density estimation) - оценка глубины ядра
# bins - кол-во прямоугольников
```

```python
sns.jointplot(x='coherence', y='firing_rate', data=dots)
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
# kind позволяет выбирать вид графика
```

```python
sns.distplot(tips['total_bill'], kde=False, bins=30)
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips, hue='sex', palette='husl')
# hue позволяет передавать столбцы с категорийными значениями
```

```python
sns.rugplot(tips['total_bill'])
```
