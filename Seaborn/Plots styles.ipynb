```python
import seaborn as sbn
```

```python
tips = sbn.load_dataset('tips')
```

```python
tips.head()
```

```python
sbn.set_style('whitegrid')

sbn.countplot(x='sex', data=tips)
sbn.despine(left=True, bottom=True)
# despine убирает линии сверху и справа 
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,4))
sbn.countplot(x='sex', data=tips)
```

```python
sbn.set_context('poster', font_scale=1)
sbn.countplot(x='sex', data=tips)
```

```python
sbn.set_context('notebook', font_scale=1)
sbn.lmplot(x='total_bill', y='tip', data=tips, hue='sex', palette='cool')
```
