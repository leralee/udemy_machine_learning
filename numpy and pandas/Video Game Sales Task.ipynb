#### 14. How many video games have "super" word in their names?

```python
# len(set([i for i in vgs['Name'] if 'super' in i.lower()]))
def super_in_name(name):
    if 'super' in name.lower().split():
        return name
    else:
        return False
```

```python
vgs['Name'].apply(lambda x: super_in_name(x))
```
