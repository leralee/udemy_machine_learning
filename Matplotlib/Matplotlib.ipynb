```python
import numpy as np
```

```python
a = np.linspace(0, 5, 10)
b = a ** 3
```

```python
a
```

```python
b
```

### Functional method

```python
plt.plot(a, b)
plt.xlabel('a array')
plt.ylabel('b array')
plt.title('plot for a and b array')
```

```python
plt.subplot(1,2,1)
plt.plot(a,b,'r')
plt.subplot(1,2,2)
plt.plot(b, a, 'b')
```

### Object oriented method

```python
figure = plt.figure()
axes = figure.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(a,b)
axes.set_xlabel('a array values')
axes.set_ylabel('b array values')
axes.set_title('Some title')
```

```python
figure1 = plt.figure()
axes1 = figure1.add_axes([0.0, 0.0, 1.0, 1.0])
axes2 = figure1.add_axes([0.3, 0.6, 1.0, 0.9])
axes1.plot(a,b)
axes2.plot(b,a)
```

## Matplotlib. Part 2

 subplot     ( )

```python
figure, axes = plt.subplots(nrows=2, ncols=4)
#       :
plt.tight_layout()
```

```python
figure, axes = plt.subplots(nrows=1, ncols=2)
plt.tight_layout()

for current_ax in axes:
    current_ax.plot(a,b)
```

```python
axes
```

```python
figure, axes = plt.subplots(nrows=1, ncols=2)
plt.tight_layout()
axes[0].plot(a,b)
axes[0].set_xlabel('a array values')
axes[0].set_ylabel('b array values')
axes[0].set_title('Some title')
axes[1].plot(b,a)
axes[1].set_xlabel('b array values')
axes[1].set_ylabel('a array values')
axes[1].set_title('Some title')
```

## Figure size, aspect ratio and dpi

```python
figure = plt.figure(figsize=(7, 2))
axes = figure.add_axes([0, 0, 1, 1])
axes.plot(a,b)
```

```python
figure, axes = plt.subplots(nrows=2, ncols=1, figsize=(7, 2))
plt.tight_layout()
axes[0].plot(a,b)
axes[1].plot(b,a)
```

```python
figure.savefig('my_plot.png')
```

```python
figure = plt.figure()
axes = figure.add_axes([0,0,1,1])
axes.plot(a,b, label = 'a,b plot')
axes.plot(b,a, label = 'b,a plot')
# label  :
axes.legend(loc=(0.1,0.1))
```

## Matplotlib. Part 3

```python
figure = plt.figure()
axes = figure.add_axes([0,0,1,1])
axes.plot(a,b, color='#10efeb', lw=5, alpha=1, ls="-.", marker='o', 
          markersize=20, markerfacecolor='red', markeredgewidth=3, markeredgecolor='purple')
#    ,    
# lw(line width) -  
# ls (line style)

```
